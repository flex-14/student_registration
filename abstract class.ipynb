```python
from abc import abstractmethod

class Animal:
    @abstractmethod
    def make_sound(self):
        pass
    def test(self):
        print("Testing////")

class Dog(Animal):
    def make_sound(self):
        print("Bark")

# Instantiate the class
dog = Dog()
dog.make_sound()# This will work fine


animal = Animal()
```

```python
from abc import ABC, abstractmethod

class Animal:
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Bark")

# Instantiate the class
dog = Dog()
dog.make_sound()  # This will work fine

# Trying to instantiate Animal will raise an error

animal = Animal()
 # Output: Can't instantiate abstract class Animal with abstract methods make_sound
```
