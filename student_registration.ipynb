```python
#using functions

def display_menu():
    print("\nWelcome to the Student Registration Portal")
    print("1. Register for UG courses")
    print("2. Register for PG courses")
    print("3. View Registered Students")
    print("4. Exit")

def select_courses(course_list):
    selected_courses = []
    while True:
        print("\nAvailable courses:")
        for i, course in enumerate(course_list, start=1):
            print(f"{i}. {course}")
        print(f"{len(course_list) + 1}. Done selecting courses")
        choice = input("Select a course by number: ")
        if choice.isdigit() and 1 <= int(choice) <= len(course_list):
            selected_courses.append(course_list[int(choice) - 1])
        elif choice == str(len(course_list) + 1):
            break
        else:
            print("Invalid choice, please try again.")
    return selected_courses

def register_student(student_id, program, courses):
    name = input("Enter your name: ")
    age = input("Enter your age: ")
    student = {
        'id': student_id,
        'name': name,
        'age': age,
        'program': program,
        'courses': courses
    }
    return student

def view_registered_students(students):
    if not students:
        print("\nNo students have registered yet.")
    else:
        print("\nRegistered Students:")
        for student in students:
            print(f"ID: {student['id']}, Name: {student['name']}, Age: {student['age']}, Program: {student['program']}, Courses: {', '.join(student['courses'])}")

def main():
    students = []
    student_id = 1
    ug_courses = ["Mathematics", "Physics", "Chemistry", "Algorithms", "Computer Science"]
    pg_courses = ["Data Science", "Cyber Security", "Machine Learning", "Artificial Intelligence", "Robotics"]

    while True:
        display_menu()
        choice = input("Enter your choice: ")
        if choice == '1':
            print("\nYou have chosen UG courses.")
            selected_courses = select_courses(ug_courses)
            if selected_courses:
                student = register_student(student_id, "UG", selected_courses)
                students.append(student)
                student_id += 1
                print(f"Student registered successfully with ID: {student['id']}")
        elif choice == '2':
            print("\nYou have chosen PG courses.")
            selected_courses = select_courses(pg_courses)
            if selected_courses: # not empty
                student = register_student(student_id, "PG", selected_courses)
                students.append(student)
                student_id += 1
                print(f"Student registered successfully with ID: {student['id']}")
        elif choice == '3':
            view_registered_students(students)
        elif choice == '4':
            print("Exiting the program")
            break
        else:
            print("Invalid choice.4")

if __name__ == "__main__":
    main()
```

```python
class Student:
    def __init__(self, student_id, name, age, program):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.program = program
        self.courses = []

    def add_course(self, course):
        if course not in self.courses:
            self.courses.append(course)

    def __str__(self):
        courses_str = ', '.join(course.name for course in self.courses)
        return f"ID: {self.student_id}, Name: {self.name}, Age: {self.age}, Program: {self.program.name}, Courses: {courses_str}"

class Program:
    def __init__(self, name):
        self.name = name
        self.courses = []
        self.students = []

    def add_course(self, course):
        if course not in self.courses:
            self.courses.append(course)

    def register_student(self, student):
        self.students.append(student)

    def __str__(self):
        return self.name

class Course:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

class RegistrationPortal:
    def __init__(self):
        self.programs = []
        self.student_id_counter = 1

    def display_menu(self):
        print("\nWelcome to the Student Registration Portal")
        print("1. Add Program")
        print("2. Add Course to Program")
        print("3. Register Student to Program")
        print("4. Select Courses for a Student")
        print("5. View Registered Students in a Program")
        print("6. View Courses for a Student in a Program")
        print("7. Exit")

    def add_program(self):
        program_name = input("Enter the name of the new program: ")
        self.programs.append(Program(program_name))
        print(f"Program '{program_name}' added successfully.")

    def add_course_to_program(self):
        program = self.select_program()
        if program:
            course_name = input("Enter the name of the new course: ")
            program.add_course(Course(course_name))
            print(f"Course '{course_name}' added to program '{program.name}' successfully.")

    def register_student_to_program(self):
        program = self.select_program()
        if program:
            name = input("Enter student's name: ")
            age = input("Enter student's age: ")
            student = Student(self.student_id_counter, name, age, program)
            program.register_student(student)
            self.student_id_counter += 1
            print(f"Student '{name}' registered to program '{program.name}' successfully with ID: {student.student_id}")

    def select_courses_for_student(self):
        student = self.select_student()
        if student:
            self.select_courses(student)

    def select_program(self):
        if not self.programs:
            print("No programs available.")
            return None
        print("\nAvailable programs:")
        for i, program in enumerate(self.programs, start=1):
            print(f"{i}. {program.name}")
        choice = input("Select a program by number: ")
        if choice.isdigit() and 1 <= int(choice) <= len(self.programs):
            return self.programs[int(choice) - 1]
        else:
            print("Invalid choice, please try again.")
            return None

    def select_student(self):
        program = self.select_program()
        if program and program.students:
            print("\nRegistered students:")
            for i, student in enumerate(program.students, start=1):
                print(f"{i}. {student.name} (ID: {student.student_id})")
            choice = input("Select a student by number: ")
            if choice.isdigit() and 1 <= int(choice) <= len(program.students):
                return program.students[int(choice) - 1]
            else:
                print("Invalid choice, please try again.")
                return None
        else:
            print("No students registered in this program.")
            return None

    def select_courses(self, student):
        program = student.program
        selected_courses = []
        while True:
            print("\nAvailable courses:")
            for i, course in enumerate(program.courses, start=1):
                print(f"{i}. {course.name}")
            print(f"{len(program.courses) + 1}. Done selecting courses")
            choice = input("Select a course by number: ")
            if choice.isdigit() and 1 <= int(choice) <= len(program.courses):
                course = program.courses[int(choice) - 1]
                student.add_course(course)
                selected_courses.append(course)
            elif choice == str(len(program.courses) + 1):
                break
            else:
                print("Invalid choice, please try again.")
        return selected_courses

    def view_registered_students_in_program(self):
        program = self.select_program()
        if program and program.students:
            print("\nRegistered Students in program:", program.name)
            for student in program.students:
                print(student)
        else:
            print("No students registered in this program.")

    def view_courses_for_student_in_program(self):
        student = self.select_student()
        if student:
            print(f"\nCourses for student {student.name} (ID: {student.student_id}):")
            for course in student.courses:
                print(course)
        else:
            print("Student not found.")

    def main(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice: ")
            if choice == '1':
                self.add_program()
            elif choice == '2':
                self.add_course_to_program()
            elif choice == '3':
                self.register_student_to_program()
            elif choice == '4':
                self.select_courses_for_student()
            elif choice == '5':
                self.view_registered_students_in_program()
            elif choice == '6':
                self.view_courses_for_student_in_program()
            elif choice == '7':
                print("Exiting the program")
                break
            else:
                print("Invalid choice, please try again.")

if __name__ == "__main__":
    portal = RegistrationPortal()
    portal.main()
```

```python
class Student:
    def set_attributes(self, student_id, name, age, program_index):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.program_index = program_index  # Store the index of the program

    def register_student_to_program(self, programs, student_id_counter):
        prog = Program()
        program = prog.select_program(programs)
        if program:
            name = input("Enter student's name: ")
            age = input("Enter student's age: ")
            student = Student()
            student.set_attributes(student_id_counter, name, age, programs.index(program))  # Store program index
            program['students'].append(student)
            print(f"Student '{name}' registered to program '{program['name']}' successfully with ID: {student.student_id}")
            return student_id_counter + 1
        return student_id_counter


class Program:
    def add_program(self, programs):
        name = input("Enter the program name: ")
        program = {'name': name, 'courses': [], 'students': []}
        programs.append(program)
        print(f"Program '{name}' added successfully.")

    def select_program(self, programs):
        print("Available programs:")
        for index, program in enumerate(programs):
            print(f"{index + 1}. {program['name']}")
        program_choice = input("Select a program by number: ")
        program_choice = int(program_choice)
        if 1 <= program_choice <= len(programs):
            return programs[program_choice - 1]
        else:
            print("Invalid choice.")
            return None

    def view_registered_students_in_program(self, programs):
        program = self.select_program(programs)
        if program and program['students']:
            print("\nRegistered Students in program:", program['name'])
            for student in program['students']:
                print(student)
        else:
            print("No students registered in this program.")


class Course:
    def add_course_to_program(self, programs):
        if not programs:
            print("No programs available. Add a program first.")
            return
        program_name = input("Enter the program name: ")
        for program in programs:
            if program['name'] == program_name:
                course_name = input("Enter the course name: ")
                program['courses'].append(course_name)
                print(f"Course '{course_name}' added to program '{program_name}' successfully.")
                return
        print("Program not found.")

    def select_student(self, programs):
        student_id = input("Enter the student ID: ")
        for program in programs:
            for student in program['students']:
                if student.student_id == int(student_id):
                    return student
        print("Student not found.")
        return None

    def select_courses_for_student(self, programs):
        student = self.select_student(programs)
        if student:
            program_index = student.program_index  # Use the stored index
            program_courses = programs[program_index]['courses']
            selected_courses = []
            while True:
                print("\nAvailable Courses:")
                for index, course in enumerate(program_courses):
                    print(f"{index + 1}. {course}")
                print(f"{len(program_courses) + 1}. Done selecting courses")  # Dynamic option for 'done'

                choice = input("Select a course or enter the option number to finish: ").strip().lower()
                if choice.isdigit():
                    course_index = int(choice) - 1
                    if 0 <= course_index < len(program_courses):
                        course_name = program_courses[course_index]
                        if course_name not in selected_courses:
                            selected_courses.append(course_name)
                            print(f"Course '{course_name}' added to selection.")
                        else:
                            print(f"Course '{course_name}' is already selected.")
                    else:
                        print("Invalid course index.")
                else:
                    print("Invalid input. Please enter a course number or option number.")

                # Check if 'done selecting' option is chosen
                if choice.isdigit() and int(choice) == len(selected_courses) + 1:
                    break

            student['courses'].extend(selected_courses)
            print(f"Selected courses for student ID {student.student_id}: {', '.join(selected_courses)}")


def display_menu():
    print("\nWelcome to the Student Registration Portal")
    print("1. Add a Program")
    print("2. Add a course to a particular program.")
    print("3. Register a student to a particular program")
    print("4. Select Courses for a Student")
    print("5. View Registered Students in a Program")


programs = []
prog = Program()
course = Course()
stud = Student()
student_id_counter = 1

while True:
    display_menu()
    choice = input("Enter your choice: ")
    if choice == '1':
        prog.add_program(programs)
    elif choice == '2':
        if not programs:
            print("No programs available. Add a program first.")
        else:
            for program in programs:
                print(f"Program: {program['name']} - Courses: {', '.join(program['courses']) if program['courses'] else 'No courses'}")
            course.add_course_to_program(programs)
    elif choice == '3':
        student_id_counter = stud.register_student_to_program(programs, student_id_counter)
    elif choice == '4':
        course.select_courses_for_student(programs)
    elif choice == '5':
        prog.view_registered_students_in_program(programs)
    else:
        print("Invalid choice.")
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.subjects = []

    def add_subject(self, subject):
        self.subjects.append(subject)

    def __repr__(self):
        return f"Student({self.name})"


class Subject:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def __repr__(self):
        return f"Subject({self.name})"


class PG:
    def __init__(self):
        self.subjects = []

    def add_subject(self, subject):
        self.subjects.append(subject)

    def __repr__(self):
        return f"PG(subjects={self.subjects})"

    def display(self):
        subjects_list = ','.join(f"{subject.name}[{','.join(student.name for student in subject.students)}]" for subject in self.subjects)
        return f"PG[{subjects_list}]"


# Creating PG instance
pg = PG()

# Creating subjects
maths = Subject('maths')
chemistry = Subject('chemistry')

# Adding subjects to PG
pg.add_subject(maths)
pg.add_subject(chemistry)

# Creating students
mubeen = Student('mubeen')
basim = Student('basim')
christa = Student('christa')

# Adding students to subjects
maths.add_student(mubeen)
maths.add_student(basim)

chemistry.add_student(christa)
chemistry.add_student(basim)

# Adding subjects to students
mubeen.add_subject(maths)
basim.add_subject(maths)
basim.add_subject(chemistry)
christa.add_subject(chemistry)

# Displaying the object references
print(repr(pg))

# Displaying the values in the desired format
print(pg.display())
```
